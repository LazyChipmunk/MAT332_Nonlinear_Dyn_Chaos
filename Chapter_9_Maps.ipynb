### One Dimensional Linear Map


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

def rec(num,a):
    return num*a


def draw_orbit(n,a,z0,ax,c):
    k=np.ones(n)
    order = 0
    orbit=[]
    orbit.append(z0)
    ax.plot(z0,0,'*b',zorder =10,markersize=14)
    ax.text(z0-0.2,-0.5, "x0",fontsize=20,zorder =10)
    i=0
    for i in k:
        order = order +1 
        z1=rec(z0,a)
        orbit.append(z1)
        ax.plot(z1,0,'*b',zorder =10,markersize=14)
        ax.arrow(z0,(order+1)*0.2,z1-z0 , 0, head_width=0.1, head_length=0.05, fc=c, ec=c, linewidth=3, zorder =10)
        z0=z1
    s= np.max(np.abs(orbit))+1
    print(s)
    ax.set_aspect("equal")
    x = np.linspace(-6, 6,100)
    y = 0*x
    ax.plot(x, y, '-k',linewidth=3)
    ax.plot(0,0,'ok',zorder =10,markersize=14)
    ax.text(0,0.2, "0",fontsize=20,zorder =10)
    return orbit

fig = plt.figure(figsize=(30,20))
gs = gridspec.GridSpec(nrows=2, ncols=3, height_ratios = [1,1], width_ratios = [1,1,1])

ax= fig.add_subplot(gs[0,0])
n=4
a=1.5
z0=0.5
draw_orbit(n,a,z0,ax,"b")
ax.set_xlim((-3, 3))
ax.set_ylim((-3, 3))
ax.set_title('a=1.5, x0=0.5',fontsize=30)

ax= fig.add_subplot(gs[0,1])
x = np.linspace(-6, 6,100)
y = 0*x
ax.plot(x, y, '-k',linewidth=3)
ax.plot(0,0,'ok',zorder =10,markersize=14)
z0=2
ax.plot(z0,0,'*b',zorder =10,markersize=14)
ax.text(0,0.2, "0",fontsize=20,zorder =10)
ax.text(z0-0.2,-0.5, "x0",fontsize=20,zorder =10)
ax.set_title('a=1, x0=2',fontsize=30)
ax.set_xlim((-3, 3))
ax.set_ylim((-3, 3))
ax.set_aspect("equal")


ax= fig.add_subplot(gs[0,2])
n=4
a=0.5
z0=2.5
ax.set_title('a=0.5, x0=2.5',fontsize=30)
ax.set_xlim((-3, 3))
ax.set_ylim((-3, 3))
draw_orbit(n,a,z0,ax,"b")


ax= fig.add_subplot(gs[1,0])
n=4
a=-1.5
z0=0.5
draw_orbit(n,a,z0,ax,"r")
ax.set_xlim((-3, 3))
ax.set_ylim((-3, 3))
ax.set_title('a=-1.5, x0=0.5',fontsize=30)

ax= fig.add_subplot(gs[1,1])
n=4
a=-1
z0=2
draw_orbit(n,a,z0,ax,"r")
ax.set_xlim((-3, 3))
ax.set_ylim((-3, 3))
ax.set_title('a=-1, x0=2',fontsize=30)


ax= fig.add_subplot(gs[1,2])
n=4
a=-0.5
z0=2.5
ax.set_title('a=-0.5, x0=2.5',fontsize=30)
draw_orbit(n,a,z0,ax,"r")
ax.set_xlim((-3, 3))
ax.set_ylim((-3, 3))
```

### Node Source

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

def rec(num, a,b,c,d):
    num1 = a*num[0] + b*num[1]
    num2 = c*num[0] + d*num[1]
    return [num1,num2]


fig = plt.figure(figsize=(30,30))
gs = gridspec.GridSpec(nrows=1, ncols=2, width_ratios = [1,1])

ax = fig.add_subplot(gs[0, 0])

a=0.8
b=0
c=0.8
d=0.2
z0=[1,2]

k=np.ones(4)
order = 0
orbit=[]
orbit.append(z0)
ax.plot(z0[0],z0[1],'-ok',zorder =10)
ax.text(z0[0]+0.1,z0[1]+0.1, str(order),fontsize=20,zorder =10)
i=0
for i in k:
    order = order +1 
    z0=rec(z0, a,b,c,d)
    orbit.append(z0)
    ax.text(z0[0]+0.1,z0[1]+0.1, str(order),fontsize=20,zorder =10)
    ax.plot(z0[0],z0[1],'-ok',zorder =10)


ax.plot(0,0,'-ok')

axsize = np.sqrt(z0[0]**2 + z0[1]**2) +2


ax.set_xlim((-axsize, axsize))
ax.set_ylim((-axsize, axsize))


x = np.linspace(-axsize, axsize,100)
y = 0*x
plt.plot(y, x, '-k',linewidth=3)
plt.plot(x, y, '-k',linewidth=3)

###eigenvector
y1 =4*x/3       
y2 = x
plt.plot(y,x, '--g',linewidth=3)    # lambda1 = 1/5
plt.plot(x,y1, '--g',linewidth=3)   # lambda2 = 4/5

ax.set_aspect("equal")
```

### Spiral 

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

def rec(num, a,b,c,d):
    num1 = a*num[0] + b*num[1]
    num2 = c*num[0] + d*num[1]
    return [num1,num2]


fig = plt.figure(figsize=(30,30))
gs = gridspec.GridSpec(nrows=1, ncols=2, width_ratios = [1,1])

ax = fig.add_subplot(gs[0, 0])

m= 6    # when m >5 it is sink when m <5 it is source
a=-1/m
b=4/m
c=-8/m
d=7/m

#Initial Condition
z0=[20,0]

k=np.ones(10)
order = 0
orbit=[]
orbit.append(z0)
ax.plot(z0[0],z0[1],'-ok',zorder =10)
ax.text(z0[0]+0.1,z0[1]+0.1, str(order),fontsize=20,zorder =10)
i=0
for i in k:
    order = order +1 
    z0=rec(z0, a,b,c,d)
    orbit.append(z0)
    ax.text(z0[0]+0.1,z0[1]+0.1, str(order),fontsize=20,zorder =10)
    ax.plot(z0[0],z0[1],'-ok',zorder =10)

ax.plot(0,0,'-ok')

axsize = 30


ax.set_xlim((-axsize, axsize))
ax.set_ylim((-axsize, axsize))

x = np.linspace(-axsize, axsize,100)
y = 0*x
plt.plot(y, x, '-k',linewidth=3)
plt.plot(x, y, '-k',linewidth=3)

ax.set_aspect("equal")
```
