```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.patches as mpatches
import math as ms

circle1 = plt.Circle((0, 0), 2, color='r',fill=False,linewidth=3)
circle2 = plt.Circle((0, 0), 1, color='g',fill=False,linewidth=3)
# 1D arrays


fig=plt.figure(figsize=(20, 20))
gs = gridspec.GridSpec(nrows=1, ncols=2, width_ratios = [1,1])

ax1=fig.add_subplot(gs[0,0])
ax1.set_xlim((-1.5, 1.5))
ax1.set_ylim((-1.5, 1.5))
t = np.arange(0,2*np.pi-0.01, 0.01)

x=1

r = x

x= r* np.cos(-t)
y= r* np.sin(-t)

ax1.plot(x,y, 'r--', linewidth=3,label='initial condition (1,0)')

x=0.8

r = x

x= r* np.cos(-t)
y= r* np.sin(-t)

ax1.plot(x,y, 'y--', linewidth=3, label='initial condition (0.8,0)')

x=1.2

r = x

x= r* np.cos(-t)
y= r* np.sin(-t)

ax1.plot(x,y, 'b--', linewidth=3,label='initial condition (1.2,0)')

x = np.arange(-2,2.01,0.01)
y = 0*x
x1=np.arange(0,2.01,0.01)
y1=0*x1

ax1.plot(x,y, 'k-', linewidth=3)
ax1.plot(y,x, 'k-', linewidth=3)
ax1.plot(x1,y1, 'g-', linewidth=6,label='Poincare Section')
ax1.text(1.3,0.1, 'x' ,fontsize=20)
ax1.text(0.1,1.3, 'y' ,fontsize=20)

ax1.arrow(1, 0, 0, -0.5, head_width=0.1, head_length=0.1, fc='k', ec='k', linewidth=3, zorder =10)
ax1.set_aspect('equal')
ax1.axis("off")
ax1.set_title('Harmonic Oscillator',fontsize=20)
ax1.legend(loc='lower right',fontsize=11)


ax1=fig.add_subplot(gs[0,1])
ax1.set_xlim((-1.5, 1.5))
ax1.set_ylim((-1.5, 1.5))


t = np.arange(0,2*np.pi-0.01, 0.01)

x=1

r = (1+np.exp(-2*t) *(x**(-2)-1))**(-0.5)

x= r* np.cos(-t)
y= r* np.sin(-t)

ax1.plot(x,y, 'r--', linewidth=3,label='initial condition (1,0)')

x=0.8

r = (1+np.exp(-0.2*t) *(x**(-2)-1))**(-0.5)

x= r* np.cos(-t)
y= r* np.sin(-t)

ax1.plot(x,y, 'y--', linewidth=3, label='initial condition (0.8,0)')

x=1.2

r = (1+np.exp(-0.2*t) *(x**(-2)-1))**(-0.5)

x= r* np.cos(-t)
y= r* np.sin(-t)

ax1.plot(x,y, 'b--', linewidth=3,label='initial condition (1.2,0)')

x = np.arange(-2,2.01,0.01)
y = 0*x
x1=np.arange(0,2.01,0.01)
y1=0*x1

ax1.plot(x,y, 'k-', linewidth=3)
ax1.plot(y,x, 'k-', linewidth=3)
ax1.plot(x1,y1, 'g-', linewidth=6,label='Poincare Section')
ax1.text(1.3,0.1, 'x' ,fontsize=20)
ax1.text(0.1,1.3, 'y' ,fontsize=20)

ax1.arrow(1, 0, 0, -0.5, head_width=0.1, head_length=0.1, fc='k', ec='k', linewidth=3, zorder =10)
ax1.set_aspect('equal')
ax1.axis("off")
ax1.set_title('Anharmonic Oscillator',fontsize=20)
ax1.legend(loc='lower right',fontsize=11)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as INT
import matplotlib.pyplot as plt
def deriv(u, t):
    dx = u[1] 
    dy = -u[0] +u[1]*(1- u[0]*u[0] )
    return [dx,dy]  
s=1.00431
u0=[s*2.0,s*0.0]

t= np.arange(0,7,0.01)
x = np.arange(-3,3,0.02)
y = np.arange(-3,3,0.02)

# Meshgrid
X,Y = np.meshgrid(x,y)

# Assign vector directions
Ex = Y 
Ey = -X +Y*(1- X*X )

fig=plt.figure(figsize=(10, 10))
ax=fig.add_subplot()
ax.streamplot(X,Y,Ex,Ey,density=3, linewidth=None)
u = INT.odeint(deriv, u0, t)
x,y= zip(*u)

ax.plot(x,y,'g',linewidth=5, label='The Periodic Orbit',zorder =10)
ax.plot(0,0,'g',linewidth=5)

x = np.arange(-3,3.01,0.01)
y = 0*x
ax.plot(x,y, 'k-', linewidth=3)
ax.plot(y,x, 'k-', linewidth=3)
ax.set_xlim((-3, 3))
ax.set_ylim((-3, 3))
ax.text(2.8,0.1, 'x' ,fontsize=20)
ax.text(-0.3,2.8, 'y' ,fontsize=20)
ax.set_aspect('equal')
ax.set_title('Van der Pol Equation',fontsize=20)
ax.legend(loc='lower right',fontsize=20)
plt.show()
```
